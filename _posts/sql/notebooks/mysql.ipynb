## Creating a table

```python
import mysql.connector
connection=mysql.connector.connect(
    host="localhost",
    user="root",
    passwd="dpm@12345", 
    database="testdbsql"
)
cursor=connection.cursor()
qry='''
CREATE TABLE IF NOT EXISTS Employee (
EmpID INTEGER PRIMARY KEY AUTO_INCREMENT,
FIRST_NAME VARCHAR (20),
LAST_NAME VARCHAR (20),
AGE INTEGER,
GENDER VARCHAR(1),
INCOME FLOAT
);
'''
try:
   cursor.execute(qry)
   print ('Table created successfully')
except:
   print ('error in creating table')
   
# connection.close()
```

# Inserting data into a table

```python
qry="""INSERT INTO Employee(
   FIRST_NAME, LAST_NAME, AGE, GENDER, INCOME)VALUES
   ('Rohan', 'Rai', 19, 'M', 40000);"""

try:
   cursor.execute(qry)
   connection.commit()
   print ('Record inserted successfully!')
except:
   connection.rollback()
   print ('Error in INSERT operation')
```

```python
qry="""INSERT INTO EMPLOYEE
      (FIRST_NAME, LAST_NAME, AGE, GENDER, INCOME)VALUES
      (%s, %s, %s, %s, %s);"""
try:
   cursor.execute(qry, ('Mina', 'Deo', 20, 'F', 60000))
   connection.commit()
   print ('Record inserted successfully')
except Exception as e:
   connection.rollback()
   print ('error in INSERT operation')
```

```python
qry="SELECT * FROM EMPLOYEE"

try:
   # Execute the SQL command
   cursor.execute(qry)
   # Fetch all the rows in a list of lists.

   results = cursor.fetchall()
   print(results); print()
   
   for col in results:
      fname = col[1]
      lname = col[2]
      age =col[3]
      gender = col[4]
      income = col[5]

      # Now print fetched result
      print ("fname={}, lname={}, age={}, gender={}, income={}".format(fname, lname, age, gender, income))
    
except Exception as e:
   print (e)
   print ("Error: unable to fecth data")
```

```python
qry="UPDATE EMPLOYEE SET INCOME = INCOME+10000 WHERE INCOME=%s"

try:
   # Execute the SQL command
   cursor.execute(qry, (40000,))
   connection.commit()
   print ("Records updated")
except Exception as e:
   print ("Error: unable to update data", e)
```

```python
qry="DELETE FROM EMPLOYEE WHERE INCOME<%s"

try:
   # Execute the SQL command
   cursor.execute(qry, (70000,))
   connection.commit()
   print ("Records deleted.")
except Exception as e:
   print ("Error: unable to delete data", e)

connection.close()
```
