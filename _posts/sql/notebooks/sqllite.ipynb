## Creating a table

```python
import sqlite3

connection = sqlite3.connect('testdb.db')
cursor = connection.cursor()

qry = '''
CREATE TABLE IF NOT EXISTS Employee(
EmpID INTEGER PRIMARY KEY AUTOINCREMENT,
FIRST_NAME TEXT (20),
LAST_NAME TEXT(20),
AGE INTEGER,
GENDER TEXT(1),
INCOME FLOAT
);
'''

try:
   cursor.execute(qry)
   print ('Table created successfully')
except:
   print ('Error in creating table')
# connection.close()
```

## Inserting data into table

```python
# connection=sqlite3.connect('testdb3.sqlite3')
# cursor=connection.cursor()

qry="""INSERT INTO Employee(
   FIRST_NAME, LAST_NAME, AGE, GENDER, INCOME)VALUES
   ('Rohan', 'Rai', 19, 'M', 40000);"""

try:
   cursor.execute(qry)
   connection.commit()
   print ('Record inserted successfully!')
except:
   connection.rollback()
   print ('Error in INSERT operation')
# connection.close()
```

## Retrieve

```python
# import sqlite3
# connection=sqlite3.connect('testdb3.sqlite3')
# cursor=connection.cursor()

qry="SELECT * FROM EMPLOYEE"

try:
   # Execute the SQL command
   cursor.execute(qry)
   # Fetch all the rows in a list of lists.

   results = cursor.fetchall()
   print(results); print()
   
   for col in results:
      fname = col[1]
      lname = col[2]
      age =col[3]
      gender = col[4]
      income = col[5]

      # Now print fetched result
      print ("fname={}, lname={}, age={}, gender={}, income={}".format(fname, lname, age, gender, income))
    
except Exception as e:
   print (e)
   print ("Error: unable to fecth data")

# connection.close()
```

## Update

```python
# import sqlite3
# connection=sqlite3.connect('testdb3.db')
# cursor=connection.cursor()

qry="UPDATE EMPLOYEE SET INCOME = INCOME+10000 WHERE INCOME=?"

try:
   # Execute the SQL command
   cursor.execute(qry, (40000,))
   connection.commit()
   print ("Records updated")
except Exception as e:
   print ("Error: unable to update data", e)
# connection.close()
```

## Delete

```python
import sqlite3
connection=sqlite3.connect('testdb3.db')
cursor=connection.cursor()


qry="DELETE FROM EMPLOYEE WHERE INCOME<?"

try:
   # Execute the SQL command
   cursor.execute(qry, (70000,))
   connection.commit()
   print ("Records deleted.")
except Exception as e:
   print ("Error: unable to delete data", e)

connection.close()
```
