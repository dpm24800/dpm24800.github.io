```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
import time

# URL to open
url = "https://github.com/dpm24800"

# Configure Chrome to run headless (no browser window)
options = Options()
options.add_argument("--headless")
options.add_argument("--disable-gpu")
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")

for i in range(20):
    # Start browser
    driver = webdriver.Chrome(ChromeDriverManager().install(), options=options)

    # Open page
    driver.get(url)
    print(f"Opened {i+1} times")

    # Wait so page can load fully (optional)
    time.sleep(1)

    # Close browser
    driver.quit()
```

```python
import requests
import time

url = "https://github.com/dpm24800"   # change this to your target URL

for i in range(30):
    try:
        response = requests.get(url, timeout=5)
        print(f"{i+1}. Status: {response.status_code}")
    except Exception as e:
        print(f"{i+1}. Error: {e}")

    time.sleep(0.1)   # optional: prevent too many requests too fast
```

```python
import requests, time

url = "https://github.com/dpm24800"   # change this

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 "
                  "(KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"
}

for i in range(30):
    try:
        response = requests.get(url, headers=headers, timeout=5)
        print(f"{i+1}. Status: {response.status_code}")
    except Exception as e:
        print(f"{i+1}. Error: {e}")

    time.sleep(0.5)
```
